```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Leer el archivo CSV
# Asegúrate de que 'notas.csv' esté en la misma carpeta que este script
df = pd.read_csv("notas_estudiantes.csv")

# Mostrar las primeras filas para verificar
print("Primeras filas del CSV:")
print(df.head())

# 2. Agrupar por asignatura y calcular la nota media
notas_medias = df.groupby('asignatura')['nota'].mean()
print("\nNotas medias por asignatura:")
print(notas_medias)

# 3. Gráfico de barras: Notas medias por asignatura
plt.figure(figsize=(10, 6))
plt.bar(notas_medias.index, notas_medias.values, color='skyblue', edgecolor='black')
plt.title('Notas Medias por Asignatura')
plt.xlabel('Asignatura')
plt.ylabel('Nota Media')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Elegir un estudiante y graficar sus notas con un gráfico de líneas
estudiante = 'Ana'
notas_estudiante = df[df['estudiante'] == estudiante]

# Verificar si el estudiante existe
if not notas_estudiante.empty:
    print(f"\nNotas del estudiante {estudiante}:")
    print(notas_estudiante)

    # Gráfico de líneas: Notas del estudiante
    plt.figure(figsize=(10, 6))
    plt.plot(notas_estudiante['asignatura'], notas_estudiante['nota'], marker='o', color='orange', linestyle='-', linewidth=2, markersize=8)
    plt.title(f'Notas de {estudiante} por Asignatura')
    plt.xlabel('Asignatura')
    plt.ylabel('Nota')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print(f"\nEl estudiante {estudiante} no se encuentra en el CSV.")
```
